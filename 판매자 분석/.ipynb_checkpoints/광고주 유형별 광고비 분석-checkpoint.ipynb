# 거래액 (현재 ~ 2년전)
- 거래액 = sum(번개송금, 번개페이, 번프) * avg(products price)


```python
# 2018-01-23 00:00:00 ~ 2020-01-23 23:59:59
start_date = '2018-01-23 00:00:00'
end_date = '2020-01-23 23:59:59'

q = f'''
select *
from (
    select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
    from (
        (select pu.uid as uid, sum(pup.pay_sum) as pay_sum, sum(pup.pay_point) as pay_point, sum(pup.pay_free) as pay_free
        from ad_power_up_point pup
        join ad_power_up pu on pup.pu_id = pu.pu_id
        where pu.created_at between '{start_date}' and '{end_date}' and pup.status <=2
        group by pu.uid)
        union all
        (select su.uid as uid, sum(sup.pay_sum) as pay_sum, sum(sup.pay_point) as pay_point, sum(sup.pay_free) as pay_free
        from ad_super_up_point sup
        join ad_super_up su on su.suid = sup.suid
        where su.create_at between '{start_date}' and '{end_date}' and su.status <= 2
        group by su.uid)
        union all
        (select sus.uid as uid, sum(susp.pay_sum) as pay_sum, sum(susp.pay_point) as pay_point, sum(susp.pay_free) as pay_free
        from ad_super_up_shop_point susp
        join ad_super_up_shop sus on susp.sus_id = sus.sus_id
        where sus.create_at between '{start_date}' and '{end_date}' and sus.status <= 2
        group by sus.uid)
        union all
        (select uid, sum(pay_point+pay_free) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_up_plus
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_today_deal
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_today_deal_plus
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_shop_popular
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
    ) as points
    group by uid) as pay
left outer join (select uid, avg(price) as avg_price, left(category_id,3) as category
                from product_info
                where create_date between '{start_date}' and '{end_date}'
                group by uid, left(category_id,3)) as product on pay.uid = product.uid;
'''
df = pd.read_sql(q, con=mysql)
```

```python
start_date = '2018-01-23 00:00:00'
end_date = '2020-01-23 23:59:59'
q = f'''
select category, sum(pay_point_sum) as pay_point          
from (
(select left(p.category_id, 3) as category, sum(pup.pay_point) as pay_point_sum
from ad_power_up_point pup
join ad_power_up pu on pup.pu_id = pu.pu_id
join product_info p on p.id = pu.pid
where pu.created_at between '{start_date}' and '{end_date}' and pup.status <= 2 and pup.pay_point != 0
group by left(p.category_id, 3))
union all
(select left(p.category_id, 3) as category, sum(sup.pay_point) as pay_point_sum
from ad_super_up_point sup
join ad_super_up su on sup.suid = su.suid
join product_info p on p.id = su.pid
where su.create_at between '{start_date}' and '{end_date}' and su.status <= 2 and sup.pay_point != 0
group by left(p.category_id, 3))
union all
(select left(p.category_id, 3) as category, sum(td.pay_point) as pay_point_sum
from ad_today_deal td
join product_info p on p.id = td.pid
where td.created_at between '{start_date}' and '{end_date}' and td.status <= 2 and td.pay_point != 0
group by left(p.category_id, 3))
union all
(select left(p.category_id, 3) as category, sum(tdp.pay_point) as pay_point_sum
from ad_today_deal_plus tdp
join product_info p on p.id = tdp.pid
where tdp.created_at between '{start_date}' and '{end_date}' and tdp.status <= 2 and tdp.pay_point != 0
group by left(p.category_id, 3))
) as ads
group by category;'''
category_point = pd.read_sql(q, con=mysql)
category_point
```

```python
# 2016-01-23 00:00:00 ~ 2018-01-23 23:59:59
start_date = '2016-01-23 00:00:00'
end_date = '2018-01-23 23:59:59'

q = f'''
select *
from (
    select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
    from (
        (select pu.uid as uid, sum(pup.pay_sum) as pay_sum, sum(pup.pay_point) as pay_point, sum(pup.pay_free) as pay_free
        from ad_power_up_point pup
        join ad_power_up pu on pup.pu_id = pu.pu_id
        where pu.created_at between '{start_date}' and '{end_date}' and pup.status <=2
        group by pu.uid)
        union all
        (select su.uid as uid, sum(sup.pay_sum) as pay_sum, sum(sup.pay_point) as pay_point, sum(sup.pay_free) as pay_free
        from ad_super_up_point sup
        join ad_super_up su on su.suid = sup.suid
        where su.create_at between '{start_date}' and '{end_date}' and su.status <= 2
        group by su.uid)
        union all
        (select sus.uid as uid, sum(susp.pay_sum) as pay_sum, sum(susp.pay_point) as pay_point, sum(susp.pay_free) as pay_free
        from ad_super_up_shop_point susp
        join ad_super_up_shop sus on susp.sus_id = sus.sus_id
        where sus.create_at between '{start_date}' and '{end_date}' and sus.status <= 2
        group by sus.uid)
        union all
        (select uid, sum(pay_point+pay_free) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_up_plus
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_today_deal
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_today_deal_plus
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
        union all
        (select uid, sum(pay_sum) as pay_sum, sum(pay_point) as pay_point, sum(pay_free) as pay_free
        from ad_shop_popular
        where created_at between '{start_date}' and '{end_date}' and status <= 2
        group by uid)
    ) as points
    group by uid) as pay
left outer join (select uid, avg(price) as avg_price, left(category_id,3) as category
                from product_info
                where create_date between '{start_date}' and '{end_date}'
                group by uid, left(category_id,3)) as product on pay.uid = product.uid;
'''
df_ = pd.read_sql(q, con=mysql) 
```

```python
df_ = df_[df_['pay_point'] != 0]
df_ = df_[['uid', 'pay_sum', 'pay_point', 'pay_free', 'avg_price', 'category']]
df_ = df_.iloc[:, 1:]
```

```python
q = f'''
select uid, sum(cnt)
from 
(
(select oi.seller_id as uid, count(om.id) as cnt
from order_mast om
join order_item oi on oi.order_mast_id = om.id
where om.create_date BETWEEN '{start_date}' AND '{end_date}' and om.order_status_cd = 'purchase_confirm'
group by oi.seller_id)
union all
(select uid, count(*) as cnt
from wire_transfer
where created_at BETWEEN '{start_date}' AND '{end_date}' and status = 'transfer_completed'
group by uid)
union all 
(select seller_uid as uid, count(*) as cnt
from bunjang_promise
where created_at BETWEEN '{start_date}' AND '{end_date}' and status = 4
group by seller_uid) 
) as trans
group by uid
'''
trans_ = pd.read_sql(q, con=redshift)
```

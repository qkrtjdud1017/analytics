```python
# 검색 많은 키워드 top 300

q = '''
SELECT lower(trim(replace(search_term, ' ', ''))) AS keyword,
       count(*) AS search_cnt
FROM bun_log_db.api_event_type_search
WHERE YEAR||MONTH||DAY = '20210303' and search_term IS NOT NULL AND search_term != ''
GROUP BY 1
ORDER BY 2 DESC
LIMIT 300
'''
top_keywords = pd.read_sql(q, con=bun_dw)
```

```python
import numpy as np
top_keywords['rank'] = np.arange(len(top_keywords))
```

```python
q = '''
WITH imp AS
  (SELECT content_id,
          lower(trim(replace(ref_term, ' ', ''))) AS ref_term,
          count(*) imp
   FROM bun_log_db.api_event_type_impression_ad
   WHERE YEAR || MONTH || DAY = '20210303'
     AND ref_source = 'sa'
     AND report = 'true'
   GROUP BY 1, 2),
     clk AS
  (SELECT content_id,
          lower(trim(replace(ref_term, ' ', ''))) AS ref_term,
          count(*) clk
   FROM bun_log_db.api_event_type_click_ad
   WHERE YEAR || MONTH || DAY = '20210303'
     AND ref_source = 'sa'
     AND report = 'true'
   GROUP BY 1, 2)
SELECT imp.content_id,
       imp.ref_term,
       imp.imp,
       CASE
           WHEN clk.clk IS NULL THEN 0
           ELSE clk.clk
       END clk,
       CASE
           WHEN clk.clk IS NULL THEN 0
           ELSE clk.clk
       END / cast(imp.imp AS decimal(10, 4)) ctr
FROM imp
LEFT JOIN clk ON clk.ref_term = imp.ref_term
AND clk.content_id = imp.content_id
WHERE imp.imp >= 100 AND ctr > 0 
'''
ctr = pd.read_sql(q, con=bun_dw)
```

```python
def get_(term):
    q = f'''
    SELECT a.content_id,
           p.name,
           a.ad_set_id,
           a.ad_set_product_id,
           a.bid_price,
           lower(trim(replace(ref_term, ' ', ''))) AS ref_term
    FROM bun_log_db.api_event_type_impression_ad a
    JOIN service1_quicket.product_info p ON a.content_id = p.id
    WHERE a.YEAR || a.MONTH || a.DAY = '20210223'
      AND a.ref_source = 'sa'
      AND a.report = 'true'
      AND lower(trim(replace(a.ref_term, ' ', ''))) = '{term}'
    GROUP BY 1, 2, 3, 4, 5, 6
    '''
    tmp = pd.read_sql(q, con=bun_dw)
    return pd.merge(many[many['ref_term'] == term], tmp, on='content_id')
```

```python
def get_data(keyword):
    enc = parse.quote(keyword)
    url = f"http://172.31.151.225:5000/api/search/uat?keyword={enc}&page=1&size=100&age=0&gender=&accuracy="
    return requests.get(url=url).json()['products']
```

```python
get_data('아이폰11')
```

```python
dataset = {}
no_output_list = []
for keyword in tqdm(top_keywords[:1000]['keyword']):
    try:
        data = get_data(keyword)
        temp = {
            'adSetId': [i['adSetId'] for i in data if i['replicated'] == False],
            'adSetProductId': [i['adSetProductId'] for i in data if i['replicated'] == False],
            'pid': [i['id'] for i in data if i['replicated'] == False],
            'bidPrice': [i['bidPrice'] for i in data if i['replicated'] == False],
            'impression': [i['impression'] for i in data if i['replicated'] == False],
            'click': [i['click'] for i in data if i['replicated'] == False],
            'ctr': [ctr[(ctr['ref_term'] == keyword) & (ctr['content_id'] == i['id'])]['ctr'].values[0] for i in data if i['replicated'] == False],
            'ecpm': [i['ecpm'] for i in data if i['replicated'] == False],
            'rank': [i['rank'] for i in data if i['replicated'] == False],
            'price': [i['price'] for i in data if i['replicated'] == False],
            'name': [i['name'] for i in data if i['replicated'] == False],
            'keywords': [i['keywords'] for i in data if i['replicated'] == False]
        }
        dataset[keyword] = temp
    except:
        no_output_list.append(keyword)
        pass
```

```python
dataset.keys()
```

```python
import json
with open('data_api.txt', 'w') as outfile:
    json.dump(dataset, outfile)
```

```python
big_dataset = {}
size = {}
for key, value in dataset.items():
    if len(value['adSetId']) >= 10:
        big_dataset[key] = pd.DataFrame.from_dict(value)
        size[key] = len(value['adSetId'])
```

```python
def get_less(keyword):
    d = big_dataset[keyword]
    return d[d['ctr'] <= d['ctr'].quantile(0.25)]
```

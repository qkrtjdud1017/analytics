```python
from dataclasses import dataclass
from typing import List


@dataclass
class Product:
    brand_id: int
    brands: List[str]
    text: str
    nfd_spans: List[List[int]]
    spans: List[List[int]]


@dataclass
class Brand:
    brand_id: int
    name_kor: str
    name_eng: str
    synonyms: List[str]
    models: List[str]
    categories: List[str]


@dataclass
class BrandDB:
    id: int
    name_kor: str
    name_eng: str
    is_active: int
    # used_for_rec: int
    # used_for_follow: int


@dataclass
class BrandDict:
    brand_id: int
    synonyms: List[str]
    models: List[str]
    categories: List[str]
```

```python
import ndjson
from typing import List


def load_data() -> List[Product]:
    file_path = './product_brand_164_ner_data.ndjson'
    json_data = _load_json_file(file_path)

    products: List[Product] = []

    for product in json_data:
        product_info = Product(
            brand_id=int(product.get('brand_id', 0)),
            brands=product.get('brands', ['']),
            text=product.get('text', ''),
            nfd_spans=product.get('nfd_spans', []),
            spans=product.get('spans', [])
        )
        products.append(product_info)

    return products


def _load_json_file(file_path: str):
    with open(file_path, 'r', encoding='utf8') as f:
        json_data = ndjson.load(f)
    return json_data
```

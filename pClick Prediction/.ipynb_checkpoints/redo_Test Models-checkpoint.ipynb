```python
df = pd.read_csv('pclick_df.csv')
```

```python
sparse = ['hour', 'imp_content_position', 'imp_page_id', 'imp_ref_page_id', 
          'imp_ref_source', 'user_sex', 'user_age', 'content_category_id']
dense = ['user_following_cnt', 'user_bunpay_count', 'owner_grade', 'owner_item_count',
         'owner_interest', 'owner_follower_cnt', 'owner_bunpay_count', 'content_price',
        'content_emergency_cnt', 'content_comment_cnt', 'content_interest', 'content_pfavcnt']
features = sparse + dense
```

```python
df = df[features + ['label']]
```

```python
for i in sparse:
    df[i] = df[i].fillna('-1')
    df[i] = preprocessing.LabelEncoder().fit_transform(df[i])
for j in dense:
    df[j] = df[j].fillna(0)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

print('train set size:', len(x_train))
print('percentage of label 1 in train set:', len([i for i in y_train if i == 1])/len(y_train))
print('percentage of label 1 in test set:', len([i for i in y_test if i == 1])/len(y_test))
```

## Logistic Regression

## Random Forest

```python
rf_result = {'class_weight': [], 'n_estimators': [], 'max_depth': [], 'label_1_prob': [], 'log_loss': [], 'accuracy': [], 'click_prob_log_loss': []}
for class_weight in range(1, 10):
    for n_estimators in range(2, 7):
        for max_depth in range(3, 7):
            rf = RandomForestClassifier(n_estimators=n_estimators,
                                        max_depth=max_depth,
                                        class_weight={0: 1, 1: class_weight},
                                        n_jobs=-1,
                                        random_state=42)
            rf.fit(x_train, y_train)
            rf_pred = rf.predict(x_test)
            rf_proba = rf.predict_proba(x_test)
            proba = []
            for p in rf_proba:
                proba.append(p[1])
            rf_result['class_weight'].append(class_weight)
            rf_result['n_estimators'].append(n_estimators)
            rf_result['max_depth'].append(max_depth)
            rf_result['label_1_prob'].append(len([i for i in rf_pred if i == 1])/len(rf_pred))
            rf_result['log_loss'].append(log_loss(y_test.astype(int), rf_pred))
            rf_result['accuracy'].append(accuracy_score(y_test.astype(int), rf_pred))
            rf_result['click_prob_log_loss'].append(log_loss(y_test.astype(int), proba))
```

```python
pd.DataFrame.from_dict(rf_result).to_csv('rf_result.csv', index=False)
```

## XGB


```python
xgb_result = {'class_weight': [], 'n_estimators': [], 'max_depth': [], 'label_1_prob': [], 'log_loss': [], 'accuracy': [], 'click_prob_log_loss': []}
for class_weight in range(1, 10):
    for n_estimators in range(2, 7):
        for max_depth in range(3, 7):
            xgb = XGBClassifier(n_estimators=n_estimators,
                               max_depth=max_depth,
                               objective='binary:logistic',
                               scale_pos_weight=class_weight,
                               n_jobs=-1,
                               random_state=42)
            xgb.fit(x_train, y_train)
            xgb_pred = xgb.predict(x_test)
            xgb_proba = xgb.predict_proba(x_test)
            proba = []
            for p in xgb_proba:
                proba.append(p[1])
            xgb_result['class_weight'].append(class_weight)
            xgb_result['n_estimators'].append(n_estimators)
            xgb_result['max_depth'].append(max_depth)
            xgb_result['label_1_prob'].append(len([i for i in xgb_pred if i == 1])/len(xgb_pred))
            xgb_result['log_loss'].append(log_loss(y_test.astype(int), xgb_pred))
            xgb_result['accuracy'].append(accuracy_score(y_test.astype(int), xgb_pred))
            xgb_result['click_prob_log_loss'].append(log_loss(y_test.astype(int), proba))
```

```python
pd.DataFrame.from_dict(xgb_result).to_csv('xgb_result.csv', index=False)
```
